```python
import sqlite3
import pandas as pd
```

```python
photo_df = pd.read_excel('photos.xlsx')
photo_list = photo_df.to_dict('records')
photo_list
```

```python
conn = sqlite3.connect('photos.sqlite')
cursor = conn.cursor()
cursor.execute("CREATE TABLE photos (id INTEGER primary key autoincrement, category_id INTEGER, filename CHAR(30) , tags CHAR(100), url CHAR(200))")
cursor.close()
conn.close()
```

```python
conn = sqlite3.connect('photos.sqlite')
cursor = conn.cursor()

for photo in photo_list:
    print(photo)
    cursor.execute("INSERT INTO photos (category_id, filename, tags, url) VALUES (?, ?, ?, ?)", (photo['category_id'], photo['filename'], photo['tags'], photo['url'].strip()))
    id = cursor.lastrowid
    print(id)
    conn.commit()
        
cursor.close()
conn.close()
```

```python
conn = sqlite3.connect('photos.sqlite')
cursor = conn.cursor()
cursor.execute("SELECT * FROM photos")
rows = cursor.fetchall()
for row in rows:
    print(row)
cursor.close()
conn.close()
```
