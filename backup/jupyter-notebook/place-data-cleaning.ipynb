```python
import pandas as pd
```

```python
df1 = pd.read_excel('place-data-tour-type.xlsx')
df2 = pd.read_excel('placelist-v.1.2.xlsx')
```

```python
df1.head(5)
```

```python
df1_dict = df1.to_dict('records')
for data in df1_dict:
    data['tour_type'] = int(data['tour_type'])
df1_dict
```

```python
for data in df1_dict:    
    query_str = "id == " + str(data['id'])
    place_by_id = df2.query(query_str)
    if len(place_by_id) > 0:
        place = place_by_id.to_dict('records')[0]
        if place['url'] is not None:
            data['url_tripadvisor'] = place['url']
        if place['address'] is not None:
            data['address'] = place['address']
        if place['phone'] is not None:
            data['phone'] = place['phone']
        if place['site'] is not None:
            data['site'] = place['site']
        if place['latitude'] is not None:
            data['latitude'] = place['latitude']
        if place['longitude'] is not None:
            data['longitude'] = place['longitude']
```

```python
df3 = pd.DataFrame(df1_dict)
df3.head(5)
```

```python
# 엑셀로 저장하기
df3.to_excel('place-data-v.1.3.xlsx')
```
